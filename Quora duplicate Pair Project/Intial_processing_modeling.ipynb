```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('train.csv')
```

```python
df
```

```python
df=df.sample(10000)
```

```python
df.dropna(inplace=True)
```

```python
question_df=df[['question1','question2']]
```

```python
question_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
questions=list(question_df['question1']) +list(question_df['question2'])
```

```python
questions
```

```python
len(questions)
```

```python
cv=CountVectorizer(max_features=3000)
```

```python
q1,q2=np.vsplit(cv.fit_transform(questions).toarray(),2)
```

```python
q1
```

```python
question_df.index
```

```python
 q1_df=pd.DataFrame(q1,index=question_df.index)
```

```python
q1_df
```

```python
q2_df=pd.DataFrame(q2,index=question_df.index)
```

```python
q2_df
```

```python
vector_df=pd.concat([q1_df,q2_df],axis=1)
```

```python
vector_df.shape
```

```python
vector_df['is_duplicate']=df['is_duplicate']
```

```python
vector_df.shape
```

```python
vector_df.head(5)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(vector_df.iloc[:,0:-1].values,vector_df.iloc[:,-1].values,test_size=0.2,random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import r2_score,accuracy_score,confusion_matrix
```

```python
model=RandomForestClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
print("accuracy is :",accuracy_score(y_pred,y_test) )
```

```python
from xgboost import XGBClassifier
```

```python
xgb=XGBClassifier()
```

```python
xgb.fit(x_train,y_train)
```

```python
y_pred1=xgb.predict(x_test)
```

```python
print("accuracy is :",accuracy_score(y_pred1,y_test) )
```
