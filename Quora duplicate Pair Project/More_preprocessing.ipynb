```python
#text preprocessing
import re
import emoji

chat_map = {
    "u": "you",
    "ur": "your",
    "r": "are",
    "pls": "please",
    "plz": "please",
    "idk": "i do not know",
    "btw": "by the way",
    "brb": "be right back",
    "dm": "direct message"
}

def text_preprocessing(text):
    text = text.lower()

    #url removal
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)

    #remove punctuation
    punctuation_pattern = r'["#,\-.@\[\\\]_{|}~]'
    text = re.sub(punctuation_pattern, '', text)

    #emoji removal
    text = emoji.demojize(text)
    text = re.sub(r':[a-z_&]+:', ' ', text)

    #chat word treatment
    text = ' '.join([chat_map.get(word, word) for word in text.split()])

    #whitespace removal
    text = re.sub(r'\s+', ' ', text)
    text = text.strip()

    return text
```

```python
question_df['question1'] = question_df['question1'].apply(text_preprocessing)
question_df['question2'] = question_df['question2'].apply(text_preprocessing)
```

```python
questions1=list(question_df['question1'])
questions2=list(question_df['question2'])
```

```python
import gensim
```

```python
from nltk import sent_tokenize
```

```python
from gensim.utils import simple_preprocess
```

```python
preprocess_question1=[]
for doc in questions1:
    raw_sen=sent_tokenize(doc)
    for sen in raw_sen:
        preprocess_question1.append(simple_preprocess(sen))
```

```python
preprocess_question2=[]
for doc in questions2:
    raw_sen=sent_tokenize(doc)
    for sen in raw_sen:
        preprocess_question2.append(simple_preprocess(sen))
```

```python
preprocess_question1
```

```python
model1=gensim.models.Word2Vec(
 window=3,
    min_count=1
)
model2=gensim.models.Word2Vec(
 window=3,
    min_count=1
)
```

```python
model1.build_vocab(preprocess_question1)
model2.build_vocab(preprocess_question2)
```

```python
model1.train(preprocess_question1,total_examples=model1.corpus_count,epochs=model1.epochs)
model2.train(preprocess_question2,total_examples=model2.corpus_count,epochs=model2.epochs)
```

```python
model1.wv.get_normed_vectors()
```

```python
y1=model1.wv.index_to_key
y2=model2.wv.index_to_key
```

```python
def document_vector1(doc):
    doc=[word for word in doc.split() if word in model1.wv.index_to_key]
    if len(doc) == 0:
        return np.zeros(model1.vector_size) 
    return np.mean(model1.wv[doc],axis=0)
```

```python
def document_vector2(doc):
    doc=[word for word in doc.split() if word in model2.wv.index_to_key]
    if len(doc) == 0:
        return np.zeros(model2.vector_size) 
    return np.mean(model2.wv[doc],axis=0)
```

```python
x1=[]
x2=[]
for doc in (question_df['question1'].values):
    x1.append(document_vector1(doc))
for doc in question_df['question2'].values:
    x2.append(document_vector2(doc))
```

```python
x1=np.array(x1)
x2=np.array(x2)
```
